```python
url="https://en.wikipedia.org/wiki/Template:2019%E2%80%9320_coronavirus_pandemic_data"
resp = requests.get(url).text
soup = BeautifulSoup(resp,'lxml')
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
df = pd.read_html(tab)[0]
df.head()
```

```python
df.columns = ['']*len(df.columns)
df.columns
```

```python
#cleaning Data
df.columns = ['Flag','Country','Confirmed_Cases','Deaths','Recovered','Ref','unnamed']
clean_df = df.drop(columns=['Flag','Ref','unnamed'])
clean_df = clean_df[0:231]
#clean_df['Country'][0].split("[")[0] # this is working - use this
clean_df['Country'] = clean_df['Country'].str.split('[',expand=True)
clean_df['Recovered'] = clean_df['Recovered'].replace(['–','-','—'],0)
clean_df.fillna(0,inplace=True)
clean_df['Country'] = clean_df['Country'].replace('China (mainland)', 'China')
clean_df=clean_df.drop([225],0)
clean_df.dropna(inplace=True)
clean_df.tail(20)
```

DATA CLEANING DONE

```python
clean_df=clean_df.astype({'Confirmed_Cases':'int64','Deaths':'int64','Recovered' : 'int64'})
clean_df.dtypes
```

```python
clean_df['Live Cases'] = clean_df['Confirmed_Cases']-clean_df['Deaths']-clean_df['Recovered']
```

```python
clean_df.sort_values(by=['Live Cases'],ascending=False,inplace=True)
clean_df.head()
```

```python
clean_df['Verdicted Cases'] = clean_df['Recovered'] + clean_df['Deaths']
clean_df['Recovery Rate'] = (clean_df['Recovered']*100) / clean_df['Verdicted Cases']
clean_df.head()
```

```python
clean_df['Pending percentage'] = (clean_df['Live Cases']*100)/ (clean_df['Confirmed_Cases'])
clean_df.head()
```

```python
clean_df['Handling value'] = clean_df['Recovery Rate'] - clean_df['Pending percentage']
clean_df.sort_values('Handling value',ascending=False,inplace=True)
#clean_df.sort_values(['Recovery Rate','Pending percentage'], ascending=[False,True],inplace=True)
clean_df[clean_df['Confirmed_Cases'] > 10000].head(10)
```

So, Above listed are the top 10 countries that crossed 10K corona cases and are handling the situation elegantly than compared to other countries.

Get Long and Lat of all countries.

```python
countries_csv = pd.read_csv('countries.csv')
countries_csv.drop('country', axis=1, inplace=True)
countries_csv.rename(columns={"name":"Country"}, inplace=True)
countries_csv.head()
```

```python
#Joining clean_df with countries_csv with country

corona_df = pd.merge(clean_df,countries_csv, on="Country")
corona_df.dropna(inplace=True)
corona_df.head(10)
```

```python
corona_df['Recovery Rate'] = round(corona_df['Recovery Rate'],2)
corona_df['Pending percentage'] = round(corona_df['Pending percentage'],2)
corona_df['Handling value'] = round(corona_df['Handling value'],2)
corona_df.head()
```

Map with Details

```python
corona_map = folium.Map(location=[0, 0], zoom_start=2, tiles="Stamen Toner")
for lat,lon,cc,d,rc,rr,pp,c in zip(corona_df['latitude'],corona_df['longitude'],corona_df['Confirmed_Cases'],corona_df['Deaths'],corona_df['Recovered'],corona_df['Recovery Rate'],corona_df['Pending percentage'],corona_df['Country']):
    label = folium.Popup('Country - '+str(c)+
                        ', Confirmed Cases - '+str(cc)+
                        ', Deaths - '+str(d)+
                        ', Recovered - '+str(rc)+
                        ', Recovery Rate %- '+str(rr)+
                        ', Pending Cases %- '+str(pp), parse_html=True)
    folium.CircleMarker(
        [lat,lon],
        radius=5,
        popup=label,
        color='#FF0000',
        fill=True,
        fill_opacity=0.6).add_to(corona_map)
corona_map
```

Using Kmeans to seperate countries into 4 clusters with the help of recovery rate calculated.

```python
kmeans_corona = KMeans(n_clusters=4, random_state=0).fit(corona_df[['Handling value']])
#kmeans_corona = KMeans(n_clusters=4, random_state=0).fit(corona_df.drop(['Live Cases','Verdicted Cases','Recovery Rate','Pending percentage','Country','latitude','longitude'], 1).dropna())
kmeans_corona.labels_
```

```python
#corona_df.drop('Cluster',1,inplace=True)
corona_df.insert(0,'Cluster',kmeans_corona.labels_)
corona_df.head()
```

Map again using clusters.

```python
colors_array = ['#FF0000','#FFFF00','#FFA500','#008000']
corona_map_clsters = folium.Map(location=[0, 0], zoom_start=2, tiles="Stamen Toner")
for lat,lon,cc,d,rc,rr,pp,c,cluster in zip(corona_df['latitude'],corona_df['longitude'],corona_df['Confirmed_Cases'],corona_df['Deaths'],corona_df['Recovered'],corona_df['Recovery Rate'],corona_df['Pending percentage'],corona_df['Country'],corona_df['Cluster']):
    label = folium.Popup('Country - '+str(c)+
                        ', Confirmed Cases - '+str(cc)+
                        ', Deaths - '+str(d)+
                        ', Recovered - '+str(rc)+
                        ', Recovery Rate %- '+str(rr)+
                        ', Pending Cases %- '+str(pp), parse_html=True)
    folium.CircleMarker(
        [lat,lon],
        radius=5,
        popup=label,
        color=colors_array[cluster],
        fill=True,
        fill_opacity=0.6).add_to(corona_map_clsters)
corona_map_clsters
```
